>**Названия и описание столбцов:**
>
>1.   `PassengerId`
    * *уникальный номер пассажира*
2.   `Survived`
    * *выживание, где* **`0`** *- не выжил, а* **`1`** *- выжил*
3.   `Pclass`
    * *класс купленного билета пассажиром*
4.   `Name`
    * *имя пассажира*
5.   `Sex`
    * *пол пассажира*
6.   `Age`
    * *возраст пассажира*
7.   `SibSp`
    * *число родственников или людей путешествующих вместе с пассажиром*
8.   `Parch`
    * *число детей/родителей путешествующих вместе с пассажиром*
9.   `Ticket`
    * *номер билета пассажира*
10.   `Fare`
    * *стоимость транспортного тарифа пассажира*
11.   `Cabin`
    * *номер каюты пассажира*
12.   `Embarked`
    * *название порта в котором пассажир сел на корабль*





>**Проверяем наличие пропусков**



```python
# пропущенных значений в age - 177, cabin - 687, embarked - 2
titanic.isnull().sum()
```

```python
# смотрим общую информацию 
titanic.info()
```

```python
# всего был 891 пассажир
titanic.describe()
```

```python
# средний возраст пассажиров был почти 30 лет, минимальный 5 месяцев, а максимальный 80 лет.
print(titanic['Age'].mean())
print(titanic['Age'].min())
print(titanic['Age'].max())
```

```python
# здесь мы видим, что из всех пассажиров выжило 342, а это 38% от общего числа пассажиров
titanic.loc[titanic['Survived'] == 1]
```

```python
# на корабле было 113 несовершеннолетних пассажиров 
titanic.loc[titanic['Age'] < 18]
```

```python
# из несовершеннолетних пассажиров выжил 61
titanic[(titanic['Survived'] == 1) & (titanic['Age'] < 18)]
```

```python
# список выживших несовершеннолетних женского пола
titanic_survived = titanic[(titanic['Survived'] == 1) & (titanic['Age'] < 18) & (titanic['Sex'] == 'female')]
titanic_survived
```

```python
# всего их 38
titanic_survived.describe()
```

```python
# самый высокий процент выживших был у пассажиров с билетами 1 класса 
sb.countplot(x = 'Pclass', hue = 'Survived', data = titanic)
```

```python
# женщин выжило больше чем мужчин
sb.countplot(x = 'Sex', hue = 'Survived', data = titanic)
```

```python
# показатели выживания пассажиров севших на корабль в разных портах(S - Southampton, C - Cherbourg, Q - Queenstown)
sb.countplot(x = 'Embarked', hue = 'Survived', data = titanic)
```
